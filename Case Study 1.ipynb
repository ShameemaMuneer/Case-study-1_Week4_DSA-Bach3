## **3. Plotting a count plot of male vs female buyers participated in the sales.**


```python
#Count Plot: Male vs Female buyers participated in the sales
plt.figure(figsize=(6,4))
sns.countplot(data['Buyer Gender'])
plt.title('Count Plot: Male vs Female buyers',fontsize=18)
plt.xlabel('Buyer Gender',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.grid()

#to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## **4. Displaying the top 5 cars based on their sales price.**

```python
#sorting the data set in Descending order of sales and assigning into a variable
sorted_data=data.sort_values('Sale Price',ascending=False).head(5)
#Reset index from 0-4
sale_sort=sorted_data.reset_index()
sale_sort
```

```python
#Finding top 5 sales 
top5_sale=sale_sort['Sale Price'].nlargest(n=5)
top5_sale
```

```python
#Displaying The top 5 car makers From the derived data 
cars_top5=sale_sort[['Make','Model','Sale Price']]
cars_top5
```

## **5. Displaying the Least 5 cars based on their Resell price.**

```python
#sorting the data set in Ascending order of Resell price and assigning into a variable
sorted_data1=data.sort_values('Resell Price').head(5)
#Reset index from 0-4
resell_sort=sorted_data1.reset_index()
resell_sort
```

```python
#Finding least 5 Resell prices
least_resell=resell_sort['Resell Price'].nsmallest(n=5)
least_resell
```

```python
#Displaying The least 5 cars based on their resell price From the derived data 
cars_least=resell_sort[['Make','Model','Resell Price']]
cars_least
```
